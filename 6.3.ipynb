```python
# Importing Libraries

import random
import string
import zlib
```

```python
# Function to return a random string of input length
def generate_random_string(length):
    return ''.join(random.choice(string.ascii_letters + string.digits) for _ in range(length))

# generating 1000 sample length of string
length_of_string = [random.randint(1000, 100000) for _ in range(1000)]

# random_string = []
# compressed_data = []
compression_ratio = []

for i in range(len(length_of_string)):
    random_string = generate_random_string(length_of_string[i])
    compressed_data = zlib.compress(random_string.encode())
    compression_ratio.append(len(random_string) / len(compressed_data))


print("Avg. Compression ratio:", sum(compression_ratio)/len(compression_ratio))
```
