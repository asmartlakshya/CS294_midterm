```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Generate a random dataset
fet = [2 ** i for i in range(13)][3:]
for a in fet:
    X, y = make_classification(n_samples=a*3, n_features=a, n_classes=2, n_informative=a-2, n_redundant=0, random_state=42, n_clusters_per_class=1)

    # Split the dataset into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

    X_new_train_full = X_train[:2*a]
    y_new_train_full = y_train[:2*a]

    X_new_train_half = X_train[:a]
    y_new_train_half = y_train[:a]
    # Initialize the kNN classifier
    k = 5  # Number of neighbors to consider
    knn = KNeighborsClassifier(n_neighbors=k)
    knn2 = KNeighborsClassifier(n_neighbors=k)

    # Train the classifier
    knn.fit(X_new_train_full, y_new_train_full)
    knn2.fit(X_new_train_half, y_new_train_half)
    # Make predictions on the test set
    y_pred = knn.predict(X_test)
    y2_pred = knn2.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    acc = accuracy_score(y_test, y2_pred)

    print("For dimensions =", a, ": Accuracy n/d = 2:", accuracy, "Accuracy n/d = 1:", acc)
```
